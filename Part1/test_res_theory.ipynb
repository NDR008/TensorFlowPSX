# Hypothesis
I originally assumed that the low-res nature of the PSX would make it easier to process, since one of the reasons Pong and other games are easy to develop AI for is that the high contrast low resolution makes it easier to have clear borders to objects.

```python
# just to find screen limits

screenXY = []

import pyautogui as py
import time
for i in range(0, 2):
    print("position", i+1)
    time.sleep(3)
    screenXY.append(py.position())
x1,y1 = screenXY[0]
x2,y2 = screenXY[1]
print(x1,y1,x2,y2)
```

```python
# Helper function for applying Canny Edge Detection

import cv2

def procImage(image):
    # convert to gray
    # processed_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    processed_img = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    # blurr a bit
    processed_img = cv2.GaussianBlur(processed_img, (5, 5), 0)
    # process for edges
    processed_img = cv2.Canny(processed_img, threshold1=100, threshold2=200)
    
    return processed_img
```

```python
# Main loop for screen grab and checking performance
# Using PIL

import numpy as np
from PIL import ImageGrab
import cv2
from time import process_time_ns, time

def screen_record_pil():
    last_time = time()
    while(True):
        printscreen = np.array(ImageGrab.grab(bbox=(x1, y1, x2, y2)))
        tmp_time = time()
        period = tmp_time - last_time
        print("time taken (ms)", period*1000,
             "resulting in a frame rate of ", 1/((period)))
        last_time = tmp_time
        cv2.imshow('window', cv2.cvtColor(printscreen, cv2.COLOR_BGR2RGB))
        if cv2.waitKey(25) & 0xFF == ord('q'):
            cv2.destroyAllWindows()
            break


screen_record_pil()
```

```python
# Main loop for screen grab and checking performance
# Using pyautogui

import numpy as np
import pyautogui
import cv2
from time import process_time_ns, time


def screen_record_py():
    last_time = time()
    while(True):
        printscreen = np.array(pyautogui.screenshot(region=(x1, y1, x2, y2)))
        tmp_time = time()
        period = tmp_time - last_time
        print("time taken (ms)", period*1000,
              "resulting in a frame rate of ", 1/((period)))
        last_time = tmp_time
        cv2.imshow('window', cv2.cvtColor(printscreen, cv2.COLOR_BGR2RGB))
        if cv2.waitKey(25) & 0xFF == ord('q'):
            cv2.destroyAllWindows()
            break


screen_record_py()
```

```python
# Main loop for screen grab and checking performance
# Using mss

import numpy as np
import mss
import cv2
from time import time


def screen_record_mss():
    last_time = time() 
    while(True):
        with mss.mss() as sct:
            bbox = (x1, y1, x2, y2)
            printscreen = np.array(sct.grab(bbox))
        tmp_time = time() 
        period = tmp_time - last_time
        print("time taken (ms)", period*1000,
              "resulting in a frame rate of ", 1/((period)))
        last_time = tmp_time
        cv2.imshow('window', printscreen)
        if cv2.waitKey(25) & 0xFF == ord('q'):
            cv2.destroyAllWindows()
            break


screen_record_mss()
```

```python
# Main loop for screen grab and checking performance
# And add Canny Edge 

import numpy as np
import mss
import cv2
from time import time


def screen_record_mss():
    last_time = time()
    while(True):
        with mss.mss() as sct:
            bbox = (x1, y1, x2, y2)
            printscreen = np.array(sct.grab(bbox))
        result = procImage(printscreen)
        tmp_time = time()
        period = tmp_time - last_time
        print("time taken (ms)", period*1000,
              "resulting in a frame rate of ", 1/((period)))
        last_time = tmp_time
        cv2.imshow('window', result)
        if cv2.waitKey(25) & 0xFF == ord('q'):
            cv2.destroyAllWindows()
            break


screen_record_mss()
```
